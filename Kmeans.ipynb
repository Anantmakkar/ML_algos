```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
x.shape
```

```python
np.unique(y)# 5 unique classes(not needed)
```

```python
#normalizing
mu=x.mean(axis=0)
std=x.std(axis=0)
x=(x-mu)/std
x=x/2
```

```python
#data visualise
plt.scatter(x[:,0],x[:,1])
```

```python
km=KMeans(n_clusters=3)
```

```python
a=km.fit_predict(x)
```

```python
for i in range(500):
    if a[i] ==0:
        plt.scatter(x[i,0],x[i,1],c='r',label='red')
    elif a[i] ==1:
        plt.scatter(x[i,0],x[i,1],c='g',label='green')
    elif a[i] ==2:
        plt.scatter(x[i,0],x[i,1],c='b',label='blue')
plt.show()
```

```python
km=KMeans(n_clusters=5)
a=km.fit_predict(x)
for i in range(500):
    if a[i] ==0:
        plt.scatter(x[i,0],x[i,1],c='r',label='red')
    elif a[i] ==1:
        plt.scatter(x[i,0],x[i,1],c='g',label='green')
    elif a[i] ==2:
        plt.scatter(x[i,0],x[i,1],c='b',label='blue')
    elif a[i] ==3:
        plt.scatter(x[i,0],x[i,1],c='c',label='black')
    elif a[i] ==4:
        plt.scatter(x[i,0],x[i,1],c='m',label='orange')

plt.show()
```

```python
sse=[]
for k in range(1,10):
    km=KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)#this function calculate sse
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('S.S.E')
K_rng=range(1,10)
plt.plot(K_rng,sse)
#from graph we can see 5 clusters are good enough
```
