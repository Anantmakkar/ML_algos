steps:
1.) load the training data
2.) find out the similarity/distance of test example with all points in the training data
3.) sort the results according to the distance/similarity scores
4.) take the mean of nearest k values for regression or take majority label from nearest k values for classification.


```python
import numpy as np
import pandas as pd   
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use("seaborn")
```

```python
x,y=make_blobs(n_samples=2000,n_features=2,cluster_std=3,centers=3,random_state=42)
#centers=3 means 3 classes class 0,1,2
n_features=x.shape[1]
m=x.shape[0]
print(x.shape,y.shape)
```

```python
xt=np.array([-10,5])#test point
for i in range(m):
    if y[i] ==0:
        plt.scatter(x[i,0],x[i,1],c='r',label='red')
    elif y[i] ==1:
        plt.scatter(x[i,0],x[i,1],c='g',label='green')
    elif y[i] ==2:
        plt.scatter(x[i,0],x[i,1],c='b',label='blue')

plt.scatter(xt[0],xt[1],color='orange',marker='*')
plt.show()
```

```python
def dist(p,q):
    return np.sqrt(np.sum((p-q)**2)) #element wise subtraction

def knn(x,y,xt,k=5):
    m=x.shape[0] #tot no of points
    dlist=[]
    for i in range(m):
        d=dist(x[i],xt)
        dlist.append((d,y[i]))
    sorteddlist=sorted(dlist)#by default sort function will compare by first attribute
    sortedlist=np.array(sorteddlist[:k])
    print(sorteddlist)
```

```python
knn(x,y,xt) # 0 label belongs to the red class and most of the nearest points belong to red class so this also belongs to red class 
```
